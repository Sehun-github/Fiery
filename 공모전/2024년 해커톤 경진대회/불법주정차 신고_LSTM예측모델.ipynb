```python
# 불필요한 칼럼 드랍(_ROAD_AD	만 남기고 드랍)
columns_to_drop = ['경도', '위도', '_PARCEL_AD', '_PARCEL_DT', '_ROAD_DT']
address = df.drop(columns=columns_to_drop)
```

```python
address
```

```python
# 민원접수일을 정수형으로 변환
address['민원접수일'] = address['민원접수일'].astype(int)
address
```

```python
# '민원접수일'과 '민원접수시간' 합치기
address['DateTime'] = address['민원접수일'].astype(str) + ' ' + address['민원접수시간']
address
```

```python
# datetime으로 변환
address['DateTime'] = pd.to_datetime(address['DateTime'], format='%Y%m%d %H:%M:%S')
address
```

```python
# 주말만 뽑아내기 2일 임시공휴일, 3일 개천절, 9일 한글날 포함
target_dates = [20231001,20231002,20231003, 20231007, 20231008,20231009, 20231014,20231015,20231021,20231022, 20231028, 20231029]
weekend = address[address['민원접수일'].isin(target_dates)]
weekend
```

```python
# 변환된 칼럼을 데이터프레임에서 제거하고 필요에 따라 재정렬
weekend = weekend.drop(columns=['민원접수일', '민원접수시간']).rename(columns={'_ROAD_AD': 'address'}).reindex(columns=['DateTime', 'address'])
weekend
```

```python
weekend[['시', '구', '도로1', '도로2', '건물번호']] = weekend['address'].str.split(' ', expand=True)
weekend
```

```python
# 주말 '중구'인 행들만 선택
weekend = weekend[weekend['구'] == '중구']
weekend
```

```python
weekend_counts = weekend.groupby('DateTime').size().reset_index(name='Count')
weekend_counts
```

```python
weekend_result = weekend_counts.groupby(pd.Grouper(key='DateTime', freq='20T')).sum().reset_index()
weekend_result
```

```python
# 이제 테스트 데이터
test_data = end_data.iloc[training_data_len - 60:, :]

# x_test랑 y_test
x_test = []
y_test = end_data.iloc[training_data_len:, :]
for i in range(60, len(test_data)):
    x_test.append(test_data.iloc[i-60:i, 1])  # 'Count' 열은 1번째 열이므로 인덱스 1을 사용

# 넘파이 배열로 변환
x_test = np.array(x_test)

# Reshape
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

# 예측값 얻어내기
predictions = model.predict(x_test)
```

```python
# 'DateTime' 열 제거
y_test_values = y_test.drop(columns=['DateTime']).values

# RMSE값
rmse = np.sqrt(np.mean(((predictions - y_test_values) ** 2)))
print(rmse)
```
