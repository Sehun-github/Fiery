```python
# 점 4개를 찍고 픽셀값 구하기

import cv2
import numpy as np

# Global variables to store the coordinates of the selected points and the image
selected_points = []
image = None

# Function to calculate the area of a quadrilateral
def area_of_quadrilateral(points):
    p1, p2, p3, p4 = points
    return 0.5 * abs(p1[0]*p2[1] + p2[0]*p3[1] + p3[0]*p4[1] + p4[0]*p1[1] - p2[0]*p1[1] - p3[0]*p2[1] - p4[0]*p3[1] - p1[0]*p4[1])

def write_text(image, text, position):
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 0.5
    font_thickness = 1
    color = (0, 255, 0)
    cv2.putText(image, text, position, font, font_scale, color, font_thickness)


# Mouse callback function
def mouse_callback(event, x, y, flags, param):
    global selected_points, image
    
    if event == cv2.EVENT_LBUTTONDOWN:
        selected_points.append((x, y))
        if len(selected_points) == 4:
            # Calculate the area of the quadrilateral
            quadrilateral_area = area_of_quadrilateral(selected_points)
            quadrilateral_area = int(quadrilateral_area)
            text_position = (selected_points[0][0], selected_points[0][1] - 10)
            write_text(image, f"pixels: {quadrilateral_area}", text_position)
            print("픽셀 값:", int(quadrilateral_area))
            # Draw lines connecting the selected points
            for i in range(len(selected_points) - 1):
                cv2.line(image, selected_points[i], selected_points[i + 1], (0, 255, 0), 2)
            cv2.line(image, selected_points[-1], selected_points[0], (0, 255, 0), 2)
            # Reset the selected points
            selected_points = []

# Load the image
image_path = r"C:\Users\admin\Desktop\공모전\2024 한국정책학회 사회문제 해커톤\흙이 있는 운동장 사진(흙만)\원본\숭의초등학교_result.png"
ff = np.fromfile(image_path, np.uint8)
image = cv2.imdecode(ff, cv2.IMREAD_UNCHANGED)

# Create a window and set the mouse callback function
cv2.namedWindow('Image')
cv2.setMouseCallback('Image', mouse_callback)

# Display the image
while True:
    cv2.imshow('Image', image)
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q'):
        image_with_title = image.copy()
        image_with_title_rgb = cv2.cvtColor(image_with_title, cv2.COLOR_RGB2BGR)
        cv2.imwrite("./images/100.png", image_with_title_rgb)
        break

cv2.destroyAllWindows()
```
