```python
# 1. 서울시 학교 기본정보를 통한 중구 초중학교 데이터 개수 확인
```

```python
import pandas as pd

# 파일 위치
file_path = r"C:\Users\admin\Desktop\공모전\2024 한국정책학회 사회문제 해커톤\데이터\데이터\서울시 학교 기본정보.csv"

df = pd.read_csv(file_path,encoding='cp949')
```

```python
# 컬럼 추출
column = ['학교종류명','설립구분','학교명','도로명주소','주야구분명']
df = df[column]
```

```python
# 도로명 주소에서 구만 가져오기
df['주소_split'] = df['도로명주소'].str.split(" ")
df['구'] = df['주소_split'].str.get(1)
```

```python
# 학교종류명이 초등학교 or 중학교
cond1 = df['학교종류명'] == '초등학교'
cond2 = df['학교종류명'] == '중학교'
cond3 = df['구'] == '중구'
df_final = df[(cond1|cond2)&cond3]
```
