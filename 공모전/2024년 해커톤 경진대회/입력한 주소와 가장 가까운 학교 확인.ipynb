```python
# 서울시 중구 초중학교 데이터 가져오기

import pandas as pd

# 파일 위치
file_path = r"C:\Users\admin\Desktop\공모전\2024 한국정책학회 사회문제 해커톤\데이터\데이터\서울시 학교 기본정보.csv"

df = pd.read_csv(file_path,encoding='cp949')


# 도로명 주소에서 구만 가져오기
df['주소_split'] = df['도로명주소'].str.split(" ")
df['구'] = df['주소_split'].str.get(1)
# 학교종류명이 초등학교 or 중학교
cond1 = df['학교종류명'] == '초등학교'
cond2 = df['학교종류명'] == '중학교'
cond3 = df['구'] == '중구'
df_final = df[(cond1|cond2)&cond3].reset_index()
column = ['학교명','도로명주소']

df_final = df_final[column]
df_final
```

```python
# 서울시 초중학교 좌표값(위도, 경도) 가져오기

import requests
from geopy.distance import geodesic
KAKAO_REST_API_KEY = '89aed7b4c8eddb94d0c753dce2e30597'  # 발급된 API 키

def kakaoAPI(address):
    """
    입력된 주소를 위도와 경도로 변환합니다.
    """
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
    headers = {"Authorization": f"KakaoAK {KAKAO_REST_API_KEY}"}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        location = response.json()['documents'][0]['address']
        lat, lng = location['y'], location['x']
        return float(lat), float(lng)
    else:
        print("주소 변환 요청에 실패했습니다.")
        return None, None

    
# 튜플을 저장할 빈 리스트 생성
schools = []

# DataFrame의 각 행을 반복합니다.
for index, row in df_final.iterrows():
    # 거리 주소를 가져옵니다.
    거리_주소 = row['도로명주소']
    
    # 카카오 API를 사용하여 위도와 경도를 가져옵니다.
    위도값, 경도값 = kakaoAPI(거리_주소)
    
    # 학교 이름, 위도 및 경도를 튜플로 묶어 리스트에 추가합니다.
    schools.append((row['학교명'], str(위도값), str(경도값)))

for school in schools:
    print(school)
```

```python
import os

# Define the directory containing the files
directory = r"C:\Users\admin\Desktop\공모전\2024 한국정책학회 사회문제 해커톤\흙이 있는 운동장 사진(흙만)\원본"

dirt_schools = []
# Function to rename files
def rename_files(directory):
    # Iterate through files in the directory
    print("흙이 있는 학교들만 추출")
    for filename in os.listdir(directory):
        # Split the filename into name and extension
        name, ext = os.path.splitext(filename)
        # Replace "_" with "" in the filename
        new_name = name.replace("result", "") + ext
        new_name = name.replace("_", "") + ext
        dirt_schools_name = new_name.split(".png")[0]
        # Construct the new file path
        for school_name, latitude, longitude in schools:
            if school_name in dirt_schools_name:
                dirt_schools.append((school_name, latitude, longitude))

# Call the function to rename files
rename_files(directory)
```

```python
# 카카오 API 사용
from geopy.distance import geodesic
import requests

KAKAO_REST_API_KEY = '89aed7b4c8eddb94d0c753dce2e30597'  # 발급된 API 키

def kakaoAPI(address):
    """
    입력된 주소를 위도와 경도로 변환합니다.
    """
    url = f'https://dapi.kakao.com/v2/local/search/address.json?query={address}'
    headers = {"Authorization": f"KakaoAK {KAKAO_REST_API_KEY}"}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        location = response.json()['documents'][0]['address']
        lat, lng = location['y'], location['x']
        return float(lat), float(lng)
    else:
        print("주소 변환 요청에 실패했습니다.")
        return None, None, None, None

def get_nearest_schools(user_address, num_schools=2):
    try:
        user_lat, user_lng = kakaoAPI(user_address)
    except Exception as e:
        print("주소가 잘못되었습니다. 다시 입력해주세요.")
        return None, None, None, None

    distances = []

    for school in dirt_schools:
        school_name, school_lat, school_lng = school
        distance = geodesic((user_lat, user_lng), (float(school_lat), float(school_lng))).kilometers
        distances.append((school_name, distance))

    # Sort distances
    distances.sort(key=lambda x: x[1])

    nearest_schools_info = []
    for i in range(min(num_schools, len(distances))):
        school_name, distance = distances[i]
        nearest_schools_info.append((school_name, distance, user_lat, user_lng))
    
    return nearest_schools_info


# 주소가 올바르게 입력될 때까지 사용자에게 주소를 입력하도록 요청
while True:
    user_address = input("주소를 입력해주세요: ")
    nearest_schools_info = get_nearest_schools(user_address)
    if nearest_schools_info != (None,None,None,None):
        for i, (school_name, distance_to_nearest_school, user_lat, user_lng) in enumerate(nearest_schools_info, start=1):
            print(f"{i}번째 가장 가까운 학교는 '{school_name}'이며, 거리는 약 {distance_to_nearest_school:.2f} 킬로미터입니다.")
        break
```

```python
def find_indices_2d(arr, value):
    indices = []
    for i, row in enumerate(arr):
        for j, elem in enumerate(row):
            if elem == value:
                indices.append((i, j))
    return indices if indices else None

nearest_school_name = nearest_schools_info[0][0]  # Assuming nearest_school_info contains (school_name, ...)
second_nearest_school_name = nearest_schools_info[1][0]  # Assuming second_nearest_school_info contains (school_name, ...)

indices_nearest = find_indices_2d(dirt_schools, nearest_school_name)
indices_nearest_2 = find_indices_2d(dirt_schools, second_nearest_school_name)

if indices_nearest:
    print(f"1번째로 가장 가까운 학교: {nearest_school_name} is located at indices: {indices_nearest}")
    print(f"2번째로 가장 가까운 학교: {second_nearest_school_name} is located at indices: {indices_nearest_2}")
else:
    print(f"The value {nearest_school_name} is not found in the array.")
```

```python
# 가장 가까운학교 경로 표시

import folium
import requests
from geopy.distance import geodesic

# 두 지점 간의 거리를 계산하고 형식화하는 함수
def calculate_distance(lat1, lng1, lat2, lng2):
    distance = geodesic((lat1, lng1), (lat2, lng2)).kilometers
    return f"거리: 약 {distance:.2f} km"

# 가장 가까운 학교에 대한 마커를 추가하는 함수
def add_marker(map_obj, school_name, school_lat, school_lng, distance_text):
    tooltip_text = f"{school_name}\n{distance_text}"
    folium.Marker(location=[school_lat, school_lng], tooltip=tooltip_text).add_to(map_obj)

        
# 사용자의 주소 위치를 중심으로 하는 지도 생성
m = folium.Map(location=[user_lat, user_lng], zoom_start=16)
    
# 사용자의 현재 위치와 가장 가까운 학교까지의 경로를 계산하는 함수
def get_route_coordinates(user_lat, user_lng, school_lat, school_lng):
    url = f"http://router.project-osrm.org/route/v1/driving/{user_lng},{user_lat};{school_lng},{school_lat}?overview=full&geometries=geojson"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        if 'routes' in data and len(data['routes']) > 0:
            # 경로의 좌표 추출
            route_coordinates = data['routes'][0]['geometry']['coordinates']
            return [(lat, lng) for lng, lat in route_coordinates]
    return None    
    
    
# 사용자의 현재 위치에 마커 추가
folium.Marker(location=[user_lat, user_lng], tooltip='현재 위치', icon=folium.Icon(color='red')).add_to(m)


# 사용자의 현재 위치와 가장 가까운 학교 사이의 경로 계산
route_coordinates = get_route_coordinates(user_lat, user_lng, float(dirt_schools[indices_nearest[0]][1]), float(dirt_schools[indices_nearest[0]][2]))

# 경로가 있을 경우 지도에 표시
if route_coordinates:
    folium.PolyLine(locations=route_coordinates, color='#003DCC', weight=5).add_to(m)
    

# 사용자의 주소에 대한 마커 추가
folium.Marker(location=[user_lat, user_lng], tooltip='현재위치', icon=folium.Icon(color='red')).add_to(m)

# 사용자의 주소와 가장 가까운 학교 사이의 거리 계산 및 형식화
distance_text = calculate_distance(user_lat, user_lng, float(dirt_schools[indices_nearest[0]][1]), float(dirt_schools[indices_nearest[0]][2]))
distance_text_2 = calculate_distance(user_lat, user_lng, float(dirt_schools[indices_nearest_2[0]][1]), float(dirt_schools[indices_nearest_2[0]][2]))


# 거리 정보를 포함하여 가장 가까운 학교에 대한 마커 추가
add_marker(m, nearest_school_name, float(dirt_schools[indices_nearest[0]][1]), float(dirt_schools[indices_nearest[0]][2]), distance_text)
add_marker(m, second_nearest_school_name, float(dirt_schools[indices_nearest_2[0]][1]), float(dirt_schools[indices_nearest_2[0]][2]), distance_text_2)
m
```
