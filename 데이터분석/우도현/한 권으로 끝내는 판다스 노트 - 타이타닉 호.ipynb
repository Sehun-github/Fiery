남자 승객의 나이의 결측치는, 남자 승객의 나이의 평균으로 채웁니다.

여자 승객의 나이의 결측치는, 여자 승객의 나이의 평균으로 채웁니다.

결측치를 모두 채운 후 age 컬럼의 평균을 출력합니다.(검증코드 실행하여 올바른 값이 출력 되는지 확인)

```python
df1 = df.copy()

female = (df1['sex'] == 'female')
male = (df1['sex'] == 'male')

female_age = df1.loc[female, 'age'].mean()
male_age = df1.loc[male, 'age'].mean()

df1.loc[female, 'age'] = df1.loc[female, 'age'].fillna(female_age)
df1.loc[male, 'age'] = df1.loc[male, 'age'].fillna(male_age)
```

```python
df1['age'].mean()
```

#### 5-8 
age를 다음과 같은 구간을 가지도록 나누어 주세요

0세 초과 ~ 15세 이하

15세 초과 ~ 30세 이하

30세 초과 ~ 45세 이하

45세 초과 ~ 최대값

나눈 다음 age_bin 컬럼을 생성하여 나눈 값을 대입하고 분포를 출력하세요

cut() 함수를 활용하세요

```python
sample = sns.load_dataset('titanic')
sample.head(3)
```

```python
sample['age_bin'] = pd.cut(sample['age'], [0, 15, 30, 45, sample['age'].max()])
sample['age_bin'].value_counts()
```

#### 5-9 나이를 3개의 균등 분할을 갖도록 구간을 나누고 분포를 출력하세요

나이가 어린 그룹부터 차례대로 'young', 'normal', 'old' 라벨을 부여하세요
qcut() 함수를 활용하세요

```python
sample['age_qbin'] = pd.qcut(sample['age'], 3, labels=['young', 'normal', 'old'])
sample['age_qbin'].value_counts()
```

#### sample 데이터프레임에 대하여 apply()를 활용하여 class 컬럼의 값을 다음과 같이 바꾸고, 분포를 출력후 변경 전과 동일한지 확인하세요

```python
sample['class'].value_counts()
```

```python
def convert(x):
    if x == 'First':
        return '일등석'
    elif x == 'Second':
        return '이등석'
    elif x == 'Third':
        return '삼등석'

sample['class'].apply(convert).value_counts()
```

### groupby()를 활용하여 다음을 출력 하세요
