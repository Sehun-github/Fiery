```python
df_seoul = df_seoul.dropna()
df_seoul.isnull().sum()
```

```python
df['평균기온_최종'] = abs(df['평균기온_제주'] - df['평균기온_서울'])
```

### 결론 - 겨울에 가는 것이 좋다.!!!!

```python
import pandas as pd
df_Nation = pd.read_csv("./data/TMAP_USE_10MONTH.csv", encoding='utf-8')
df_Nation.head()
```

```python
len(df_Nation)
```

```python
groups = df_Nation.groupby("지역(시도)")
groups.size()
```

```python
groups_2 = df_Nation.groupby("검색지유형1")
groups_2.size()
```

```python
groups_3 = df_Nation.groupby("검색지유형2")
len(groups_3.size())
```

```python
def conditions(row):
    if (row['검색지랭킹(시군구내)'] == 1):
        val = 5
    elif (row['검색지랭킹(시군구내)'] == 2):
        val = 4
    elif (row['검색지랭킹(시군구내)'] == 3):
        val = 3
    elif (row['검색지랭킹(시군구내)'] == 4):
        val = 2
    elif (row['검색지랭킹(시군구내)'] == 5):
        val = 1
    else:
        val = 0
    return val
```

```python
df_Nation['value']= df_Nation.apply(conditions, axis=1)
```

```python
df_Nation.head()
```

```python
df = df_Nation.groupby(['지역(시도)','검색지유형1'])[['value']].sum()
df
```

```python
df_sort_group_top3 = df.sort_values(by="value", ascending=False).groupby("지역(시도)").head(3)
df_row = df_sort_group_top3.sort_values(by=["지역(시도)", "value"], ascending=[True, False])
df_row
```
