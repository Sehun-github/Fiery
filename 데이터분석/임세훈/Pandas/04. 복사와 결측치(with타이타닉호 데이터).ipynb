## 결측치 확인 - isnull(), isnan()
* 결측치는 비어있는 데이터를 의미합니다.

```python
# isnull()
df.isnull().sum()
```

```python
# isna()
df.isna().sum()
```

```python
# 전체 결측 데이터의 갯수를 합산
df.isnull().sum().sum()
```

```python
# isnull <-> notnull
df.notnull().sum()
```

```python
# null 행만 출력
df.loc[df['age'].isnull()]
```

## NaN 값이 있는 데이터 제거하기 (dropna)
* 기본 옵션 값은 how=any로 설정되어 있으며, 다음과 같이 변경할 수 있습니다.
* any: 1개 라도 NaN값이 존재시 drop //// 기본값
* all: 모두 NaN값이 존재시 drop

```python
df1 = df.copy()
df1.tail()
```

```python
df1.dropna() # df1.dropna(how='any')
```

```python
df1.dropna(how='all')
```
