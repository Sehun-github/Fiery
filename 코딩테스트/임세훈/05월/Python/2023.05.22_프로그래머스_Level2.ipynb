```python
# 타겟 넘버 - 너비 우선 탐색(BFS)

def solution(numbers, target):
    number_list = [0]
    for i in numbers:
        sub = []
        for j in number_list:
            sub.append(j+i)
            sub.append(j-i)
        number_list = sub
    return number_list.count(target)

aa = [1, 1, 1, 1, 1]
bb= 3
print(solution(aa,bb))



# 다른사람 풀이 - 재귀함수 사용
def solution(numbers, target):
    if not numbers and target == 0 :
        return 1
    elif not numbers:
        return 0
    else:
        return solution(numbers[1:], target-numbers[0]) + solution(numbers[1:], target+numbers[0])
```

```python
# k진수에서 소수 개수 구하기 - 1차시도 1/3 실패

def solution(n, k):
    
    # n값을 k진수로 변환
    value = ""
    while n:
        value += str(n % k)
        n = n // k
    
    value = value[::-1]

    num_list = str(value).split("0") # 0으로 split
    
    # 띄어쓰기 "", "1" 없애기
    num2_list = num_list.copy()
    for i in range(len(num_list)):
        if num_list[i] == "" or num_list[i] == "1":
            num2_list.pop(i)

    num_list = list(map(int,num2_list))
    
    # 소수 구하기
    cnt=0
    for i in num_list:
        for j in range(2,i):
            if i% j == 0:
                break
        cnt +=1
    
    return cnt
    


print(solution(437674,3))


## 2차 시도 - 테스트 케이스 1번 시간초과 실패 빼고 전부 통과 (띄어쓰기와 1일때 삭제하는 로직을 뒤에서 부터 삭제 & 소수구하기 로직 수정)

def solution(n, k):
    
    # n값을 k진수로 변환
    value = ""
    while n:
        value += str(n % k)
        n = n // k
    
    value = value[::-1]

    num_list = str(value).split("0") # 0으로 split
      
    # 띄어쓰기 "", "1" 없애기
    num2_list = num_list.copy()  
    for i in range(len(num_list)-1,-1,-1):
        if num_list[i] == "" or num_list[i] == "1":
            num2_list.pop(i)
    
    num_list = list(map(int,num2_list))
    
    # 소수 구하기
    answer=0
    for i in num_list:
        cnt=0
        for j in range(2,i):
            if i% j == 0:
                cnt +=1
                break
        if cnt == 0:
            answer +=1
    
    return answer


## 3차시도 - 소수 구하는 공식이 n-1까지가 아니라 루트 n까지구하는 것도 가능하다는 것을 알게 됨

def solution(n, k):
    
    if n==1:
        return 0
    else:
        pass
    
    # n값을 k진수로 변환
    value = ""
    while n:
        value += str(n % k)
        n = n // k
    
    value = value[::-1]

    num_list = str(value).split("0") # 0으로 split
    
    # 띄어쓰기 "", "1" 없애기
    num2_list = num_list.copy()  
    for i in range(len(num_list)-1,-1,-1):
        if num_list[i] == "" or num_list[i] == "1":
            num2_list.pop(i)
    
    num_list = list(map(int,num2_list))
    
    # 소수 구하기
    answer=0
    for i in num_list:
        cnt=0
        for j in range(2,int(i**0.5)+1):
            if i% j == 0:
                cnt +=1
                break
        if cnt == 0:
            answer +=1
    
    return answer
```

```python
# [3차] 압축

def solution(msg):
    alpha = {}
    # A~Z까지 숫자 할당
    for i in range(65,91):
        alpha[chr(i)] = i-64
    
    
    cnt = 0
    answer = []
    n = len(msg)
    while n != 0:
        if msg[0:2+cnt] not in alpha:
            alpha[msg[0:2+cnt]] = max(alpha.values())+1
            answer.append(alpha[msg[0:1+cnt]])
            msg = msg[cnt+1:]
            cnt = 0
        else:
            cnt +=1
        n -= 1
        
    answer.append(alpha[msg]) # 마지막 남은 msg 값 추가
    return answer


aa = "TOBEORNOTTOBEORTOBEORNOT"
print(solution(aa))
```
