```python
# 뒤에 있는 큰 수 찾기 - 73.9점 (시간초과)

def solution(numbers):
    answer = []
    for i in range(len(numbers)):
        for j in range(i+1, len(numbers)):
            if numbers[j] > numbers[i]:
                answer.append(numbers[j])
                break
            elif j == len(numbers)-1:
                answer.append(-1)
                break
            
    answer.append(-1)
    
    return answer

aa = [9, 1, 5, 3, 6, 2]

print(solution(aa))


# 두번째 시도 - 미리 answer에 0을 할당 (똑같이 73.9점)
def solution(numbers):
    answer = [0 for _ in range(len(numbers))] 
    for i in range(len(numbers)):
        for j in range(i+1, len(numbers)):
            if numbers[j] > numbers[i]:
                answer[i] = numbers[j]
                break
            elif j == len(numbers)-1:
                answer[i] = -1
                break
                
    answer[-1] = -1
    
    return answer

aa = [9, 1, 5, 3, 6, 2]

print(solution(aa))

# 3번쨰 시도 - -1을 미리 할당 (82.6점)
def solution(numbers):
    answer = [-1 for _ in range(len(numbers))] 
    for i in range(len(numbers)):
        for j in range(i+1, len(numbers)):
            if numbers[j] > numbers[i]:
                answer[i] = numbers[j]
                break
    
    return answer

aa = [9, 1, 5, 3, 6, 2]

print(solution(aa))

## 4번째 시도 - Stack 알고리즘 활용
def solution(numbers):
    answer = [-1 for _ in range(len(numbers))] 
    stack = []
    for i in range(len(numbers)):
        while stack and numbers[stack[-1]] < numbers[i]:
            answer[stack.pop()]= numbers[i]
        stack.append(i)
    
    return answer

aa = [9, 1, 5, 3, 6, 2]

print(solution(aa))
```

```python
# 2 x n 타일링 - 피보내치 수열

def solution(n):
    answer = [0 for _ in range(n)]
    answer[0],answer[1] = 1,2
    
    for i in range(2,n):
        answer[i] = (answer[i-2] + answer[i-1]) % 1000000007
    
    return answer[n-1]

print(solution(100))


# 리스트 안쓰고 풀기
def solution(n):
    a, b = 1, 1
    for i in range(1, n):
        a, b = b, (a + b) % 1000000007
    return b
print(solution(100))
```

```python
# 2개 이하로 다른 비트 - 81.8점 (시간초과)

def solution(numbers):
    answer = [0] * len(numbers)
    
    for i in range(len(numbers)):
        x = numbers[i]
        y = x+1
        while True:        
            bin_x = bin(x)[2:]
            bin_y = bin(y)[2:]
            while len(bin_y) != len(bin_x):
                bin_x = "0" + bin_x
            cnt = 0
            for j in range(len(bin_x)):
                if bin_x[j] != bin_y[j]:
                    cnt +=1
                
            if cnt == 1 or cnt ==2:
                break           
            y+=1
        answer[i] = y
      
    return answer

aa = [2,15]
print(solution(aa))


## 성공(풀이보고 품) - 2진수에서 1을 더하면 0부분에서 1로 바뀌고 다음부분이 1에서 0으로 바뀌어야함
def solution(numbers):
    answer = [0] * len(numbers)
    
    for i in range(len(numbers)):
        if numbers[i] % 2 == 0:
            answer[i] = numbers[i]+1
    
        else:
            number_bin = "0"+ bin(numbers[i])[2:]
            number_bin = number_bin[:number_bin.rindex('0')]+"10" +number_bin[number_bin.rindex('0')+2:]
            answer[i] = int(number_bin,2)
    return answer
```

```python
# 숫자 변환하기 - 56.3점

def solution(x, y, n):
    answer = [x]
    cnt = 0
    while True:
        if y in answer:
            return cnt
        else:
            for i in range(len(answer)):
                number = answer.pop(i)
                answer.append(number*2)
                answer.append(number*3)
                answer.append(number+n)
            cnt +=1

        if min(answer) > y: 
            return -1

print(solution(10,40,30))


# deque를 통해 시간 단축 - 75점 (나머지 시간초과)
from collections import deque
def solution(x, y, n):
    answer = deque([x])
    cnt = 0
    while True:
        if y in answer:
            return cnt
        else:
            for i in range(len(answer)):
                number = answer.popleft()
                answer.append(number*2)
                answer.append(number*3)
                answer.append(number+n)
            cnt +=1
        if min(answer) > y: 
            return -1
        
print(solution(10,40,5))


# set 을 통한 시간 단축 - 성공
from collections import deque
def solution(x, y, n):
    answer = set()
    answer.add(x)
    cnt = 0
    while answer:
        if y == max(answer):
            return cnt
        else:
            s = set()
            for i in answer:
                number = i
                if number*2 <=y:
                    s.add(number*2)
                if number*3 <=y:
                    s.add(number*3)
                if number+n <=y:
                    s.add(number+n)
            cnt +=1
            answer = s
    return -1

print(solution(10,40,5))
```
