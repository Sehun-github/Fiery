```python
# 두 큐 합 같게 만들기 - 시간초과(93.3점)

from collections import deque
def solution(queue1, queue2):
    queue1 = deque(queue1)
    queue2 = deque(queue2)
    sum_queue1 = sum(queue1)
    sum_queue2 = sum(queue2)
    
    cnt = 0
    while queue1 and queue2 and sum_queue1 != sum_queue2:
        if sum_queue1 > sum_queue2:
            queue1_popleft = queue1.popleft()
            queue2.append(queue1_popleft)
            
            sum_queue1 -= queue1_popleft
            sum_queue2 += queue1_popleft
        else:
            queue2_popleft = queue2.popleft()
            queue1.append(queue2_popleft)
            sum_queue1 += queue2_popleft
            sum_queue2 -= queue2_popleft
        cnt += 1
        
    if queue1==deque([]) or queue2==deque([]):
        cnt = -1
        
    return cnt

aa = [3, 2, 7, 2]
bb = [4, 6, 5, 1]
print(solution(aa,bb))


# 두번째 시도 - 성공(같아지진 않고 계속 왔다갔다 하는 경우가 생길수 있음, 따라서 max_count를 할당하고 break를 작동하게 로직구현)
from collections import deque
def solution(queue1, queue2):
    queue1 = deque(queue1)
    queue2 = deque(queue2)
    sum_queue1 = sum(queue1)
    sum_queue2 = sum(queue2)
    max_count = len(queue1) * 4
    
    if (sum_queue1+sum_queue2) % 2 != 0:
        return -1
    
    cnt = 0
    while queue1 and queue2 and sum_queue1 != sum_queue2:
        if sum_queue1 > sum_queue2:
            queue1_popleft = queue1.popleft()
            queue2.append(queue1_popleft)
            
            sum_queue1 -= queue1_popleft
            sum_queue2 += queue1_popleft
        else:
            queue2_popleft = queue2.popleft()
            queue1.append(queue2_popleft)
            sum_queue1 += queue2_popleft
            sum_queue2 -= queue2_popleft
        cnt += 1
        
        if cnt > max_count:
            return -1
        
    if queue1==deque([]) or queue2==deque([]):
        cnt = -1
        
    return cnt
```

```python
# 메뉴 리뉴얼 - 보고 품

from itertools import combinations
from collections import Counter


def solution(orders, course):
    answer = []
    for c in course:
        temp = []
        for order in orders:
            combi = combinations(sorted(order), c)
            temp += combi
        counter = Counter(temp)
        
        if len(counter) != 0 and max(counter.values()) != 1:
            answer += [''.join(f) for f in counter if counter[f] == max(counter.values())]

    return sorted(answer)

aa = ["ABCFG", "AC", "CDE", "ACDE", "BCFG", "ACDEH"]
bb = [2,3,4]

print(solution(aa,bb))
```
