```python
# 명예의 전당 (1)

def solution(k, score):
    answer = []
    score_list = []
    for i in score:
        if score_list == [] or len(score_list) < k:
            score_list.append(i)
        else:
            if i > min(score_list):
                score_list.pop(score_list.index(min(score_list)))
                score_list.append(i)
        
        answer.append(min(score_list))
    return answer
```

```python
# 실패율

def solution(N, stages):
    answer = {}
    result = []
    fail_stage = len(stages)
    for i in range(1,N+1):
        answer[i] = stages.count(i)/fail_stage
        fail_stage -= stages.count(i)
        if fail_stage == 0:
            for j in range(i+1,N+1):
                answer[j] = 0
            break
    aa = sorted(answer.items(), key=lambda x:(-x[1]))

    for k,v in aa:
        result.append(int(k))   
    return result
```

```python
# 체육복 - 67점

def solution(n, lost, reserve):
    answer = 0
    check = [True] * n
    
    a = list(set(reserve) - set(lost))
    b = list(set(lost) - set(reserve))
    

    for i in b:
        check[i-1] = False
        
    while a:
        index = a.pop(0)
        index -= 1
        index_1 = index + 1
        index_2 = index - 1
    
        if index_1 < n:
            if check[index_1] == False:
                check[index_1] = True
                
        elif index_2 > -1: 
            if check[index_2] == False:
                check[index_2] = True
    return sum(check)

print(solution(5,[2, 4],[1, 3, 5]))

# 풀이 보고 품 - set으로 해결
def solution(n, lost, reserve):
    answer = 0
    
    lost_reserve = set(lost) - set(reserve)
    reserve_lost = set(reserve) - set(lost)
    
    for i in reserve_lost:
        if i-1 in lost_reserve:
            lost_reserve.remove(i-1)
        elif i+1 in lost_reserve:
            lost_reserve.remove(i+1)
    
    return n-len(lost_reserve)
```

```python
# 대충 만든 자판

def solution(keymap, targets):
    answer = []
    for target in targets:
        sum = 0
        for s in target:
            result = []
            for key in keymap:
                if s in key:
                    result.append(key.index(s)+1)
            
            if result != []: 
                sum += min(result)
            
            if result == []:
                sum = 0
                break
        if sum == 0:
            sum = -1
        answer.append(sum)
    return answer
```

```python
# 둘만의 암호

def solution(s, skip, index):
    answer = ''
    for alphabet in s:
        i = index
        while i>0:
            alphabet = chr(ord(alphabet) + 1) 
            if ord(alphabet) > ord('z'):
                alphabet = chr(ord(alphabet) - 26)
            if alphabet not in skip:
                i -= 1
            
        answer += alphabet
    return answer
aa = "aukks"
bb = "wbqdx"
cc = 5
print(solution(aa,bb,cc))
```
